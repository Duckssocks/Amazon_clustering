```python
import pandas as pd
df = pd.read_csv(r'C:\study\research_team\Amazon Customer Behavior Survey.csv', index_col=0)
```

```python
df.columns
```

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"열 '{column}'의 고윳값")
    print(unique_values)
    print()
```

```python
col_list = ['age', 'Gender', 'Purchase_Frequency', 'Purchase_Categories', 'Personalized_Recommendation_Frequency',
            'Browsing_Frequency', 'Personalized_Recommendation_Frequency ', 'Recommendation_Helpfulness',
            'Shopping_Satisfaction', 'Improvement_Areas']
```

```python
df2 = df[col_list]
df2.head()
```

```python
'''age 3인거 제거'''
df3 = df2[df2['age']>3]
```

```python
'''Gender 정리 남, 여, 공개x'''
df3['Gender_Male'] = df3['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
df3['Gender_Female'] = df3['Gender'].apply(lambda x: 1 if x == 'Female' else 0)
df3['Gender_None'] = df3['Gender'].apply(lambda x: 1 if x in ['Prefer not to say', 'Others'] else 0)
df3[['Gender_Male', 'Gender_Female', 'Gender_None']].describe()
df4 = df3[['age', 'Gender_Male', 'Gender_Female', 'Gender_None', 'Purchase_Frequency', 'Purchase_Categories', 'Personalized_Recommendation_Frequency',
            'Browsing_Frequency', 'Personalized_Recommendation_Frequency ', 'Recommendation_Helpfulness',
            'Shopping_Satisfaction', 'Improvement_Areas']]
df4.info()
```

```python
'''구매품목 인코딩'''
categories = ['Beauty and Personal Care', 'Clothing and Fashion', 'Groceries and Gourmet Food', 'Home and Kitchen', 'Others']

for category in categories:
    df4['Purchase_Categories_'+category.replace(" ", "_")] = df4['Purchase_Categories'].apply(lambda x: 1 if category in x.split(';') else 0)

df4[['Purchase_Categories_Beauty_and_Personal_Care','Purchase_Categories_Clothing_and_Fashion','Purchase_Categories_Groceries_and_Gourmet_Food','Purchase_Categories_Home_and_Kitchen','Purchase_Categories_Others']].describe()
df5 = df4[['age', 'Gender_Male', 'Gender_Female', 'Gender_None', 
            'Purchase_Categories_Beauty_and_Personal_Care','Purchase_Categories_Clothing_and_Fashion','Purchase_Categories_Groceries_and_Gourmet_Food','Purchase_Categories_Home_and_Kitchen',
            'Purchase_Frequency', 'Personalized_Recommendation_Frequency',
            'Browsing_Frequency', 'Personalized_Recommendation_Frequency ', 'Recommendation_Helpfulness',
            'Shopping_Satisfaction', 'Improvement_Areas']]
df5.info()
```

```python
df5['Purchase_Frequency']
def assign_number(value):
    if value == 'Multiple times a week':
        return 5
    elif value == 'Once a week':
        return 4
    elif value == 'Few times a month':
        return 3
    elif value == 'Once a month':
        return 2
    elif value == 'Less than once a month':
        return 1
    else:
        return None

# 숫자 할당하여 새로운 열 생성
df5['Purchase_Frequency'] = df5['Purchase_Frequency'].apply(assign_number)
df6 = df5[['age', 'Gender_Male', 'Gender_Female', 'Gender_None', 
            'Purchase_Categories_Beauty_and_Personal_Care','Purchase_Categories_Clothing_and_Fashion','Purchase_Categories_Groceries_and_Gourmet_Food','Purchase_Categories_Home_and_Kitchen',
            'Purchase_Frequency', 'Browsing_Frequency', 
            'Personalized_Recommendation_Frequency','Personalized_Recommendation_Frequency ', 'Recommendation_Helpfulness',
            'Shopping_Satisfaction', 'Improvement_Areas']]
```

```python
df6['Browsing_Frequency']
def assign_number(value):
    if value == 'Multiple times a day':
        return 4
    elif value == 'Few times a week':
        return 3
    elif value == 'Few times a month':
        return 2
    elif value == 'Rarely':
        return 1
    else:
        return None

# 숫자 할당하여 새로운 열 생성
df6['Browsing_Frequency'] = df6['Browsing_Frequency'].apply(assign_number)
df7 = df6[['age', 'Gender_Male', 'Gender_Female', 'Gender_None', 
            'Purchase_Categories_Beauty_and_Personal_Care','Purchase_Categories_Clothing_and_Fashion','Purchase_Categories_Groceries_and_Gourmet_Food','Purchase_Categories_Home_and_Kitchen',
            'Purchase_Frequency', 'Browsing_Frequency', 
            'Personalized_Recommendation_Frequency','Personalized_Recommendation_Frequency ', 'Recommendation_Helpfulness',
            'Shopping_Satisfaction', 'Improvement_Areas']]
```

```python
df7['Personalized_Recommendation_Frequency']
df7['Personalized_Recommendation_Frequency ']
def assign_number(value):
    if value == 'Yes':
        return 3
    elif value == 'Sometimes':
        return 2
    elif value == 'No':
        return 1
    else:
        return None

df7['Personalized_Recommendation_Frequency_Purchase'] = df7['Personalized_Recommendation_Frequency'].apply(assign_number)
df7['Personalized_Recommendation_Frequency'] = df7['Personalized_Recommendation_Frequency ']
df8 = df7[['age', 'Gender_Male', 'Gender_Female', 'Gender_None', 
            'Purchase_Categories_Beauty_and_Personal_Care','Purchase_Categories_Clothing_and_Fashion','Purchase_Categories_Groceries_and_Gourmet_Food','Purchase_Categories_Home_and_Kitchen',
            'Purchase_Frequency', 'Browsing_Frequency', 
            'Personalized_Recommendation_Frequency_Purchase','Personalized_Recommendation_Frequency', 'Recommendation_Helpfulness',
            'Shopping_Satisfaction', 'Improvement_Areas']]
```

```python
df8['Recommendation_Helpfulness']
def assign_number(value):
    if value == 'Yes':
        return 3
    elif value == 'Sometimes':
        return 2
    elif value == 'No':
        return 1
    else:
        return None
    
df8['Recommendation_Helpfulness'] = df8['Recommendation_Helpfulness'].apply(assign_number)
df9=df8[['age', 'Gender_Male', 'Gender_Female', 'Gender_None', 
            'Purchase_Categories_Beauty_and_Personal_Care','Purchase_Categories_Clothing_and_Fashion','Purchase_Categories_Groceries_and_Gourmet_Food','Purchase_Categories_Home_and_Kitchen',
            'Purchase_Frequency', 'Browsing_Frequency', 
            'Personalized_Recommendation_Frequency_Purchase','Personalized_Recommendation_Frequency', 'Recommendation_Helpfulness',
            'Shopping_Satisfaction', 'Improvement_Areas']]
```
